```python
import matplotlib.pyplot as plt
import csv
from datetime import datetime

x = []
y = []

with open('data_20191013.csv','r') as csvfile:
    plots = csv.DictReader(csvfile, delimiter=',')
    for row in plots:
        #t = datetime.strptime(row["time"], "%Y-%m-%dT%H:%M:%S.%f")
        t = row["time"][:16]
        #print(t)
        x.append(t)
        y.append(float(row["temperature"]))


def plot(x, y, ax, title, y_label):
    ax.set_title(title)
    ax.set_ylabel(y_label)
    ax.plot(x, y)
    ax.margins(x=0, y=0)
    
    xloc = plt.MaxNLocator(20)
    ax.xaxis.set_major_locator(xloc)
    yloc = plt.MaxNLocator(10)
    ax.yaxis.set_major_locator(yloc)
    ax.tick_params(axis='x', labelrotation=45)
    ax.set_ylim(bottom=15)
    ax.set_ylim(top=30)

fig, ax = plt.subplots(figsize=(20,10))
plot(x, y, ax, 'Data', 'Temperature')
```
